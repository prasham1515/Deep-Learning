```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import accuracy_score
```

```python
df = load_iris()
x = pd.DataFrame(df.data, columns = df.feature_names)
```

```python
y=pd.DataFrame(df.target)
```

```python
x.plot(kind="scatter", x="sepal length (cm)",   y="sepal width (cm)")
plt.show()
```

```python
x.plot(kind="scatter", x="petal length (cm)",   y="petal width (cm)")
plt.show()
```

```python
x1 = x[['petal length (cm)', 'petal width (cm)']]
x2 = x[['sepal length (cm)', 'sepal width (cm)']]
```

```python
x1=x1.values
x2=x2.values
```

```python
from sklearn.mixture import GaussianMixture
```

```python
gmm = GaussianMixture(n_components=3).fit(x1)
pred1 = gmm.predict(x1)
plt.scatter(x1[:, 0], x1[:, 1], c=pred1)
plt.legend()
print(silhouette_score(x1, pred1))
```

```python
gmm = GaussianMixture(n_components=3).fit(x2)
pred2 = gmm.predict(x2)
plt.scatter(x2[:, 0], x2[:, 1], c=pred2)
plt.legend()
print(silhouette_score(x2, pred2))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=3).fit(x1)
pred3 = ac.fit_predict(x1)
plt.scatter(x1[:, 0], x1[:, 1], c=pred3)
plt.legend()
print(silhouette_score(x1, pred3))
```

```python
ac = AgglomerativeClustering(n_clusters=3).fit(x2)
pred4 = ac.fit_predict(x2)
plt.scatter(x2[:, 0], x2[:, 1], c=pred4)
plt.legend()
print(silhouette_score(x2, pred4))
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

links = linkage(x1, 'single')
dendrogram(links, orientation='top', distance_sort='descending', show_leaf_counts=True)

plt.show()
```

```python
links = linkage(x2, 'single')
dendrogram(links, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
pred5 = kmeans.fit_predict(x1)
plt.scatter(x1[:, 0], x1[:, 1], c=pred5)
plt.scatter(kmeans.cluster_centers_[:,   0], kmeans.cluster_centers_[:,1],c = 'red')
plt.legend()
print(silhouette_score(x1, pred5))
```

```python
kmeans = KMeans(n_clusters=3)
pred6 = kmeans.fit_predict(x2)
plt.scatter(x2[:, 0], x2[:, 1], c=pred6)
plt.scatter(kmeans.cluster_centers_[:,   0], kmeans.cluster_centers_[:,1],c = 'red')
plt.legend()
print(silhouette_score(x2, pred6))
```
