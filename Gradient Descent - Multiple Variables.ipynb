# Applying the gradient descent algorithm for multiple variables



## Importing Modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Defining Functions

### * Mean Squared Error
### * Gradient Descent
### * Normalization

```python
def mean_squared_error(y_true, y_predicted):
    cost = np.sum((y_true-y_predicted)**2) / len(y_true)
    return cost

def gradient_descent(a,b,c,d,e ,y, iterations = 1000, learning_rate = 0.0001,stopping_threshold = 1e-6):
    
    np.random.seed(10)
    
    w1 = np.random.normal(scale=0.5)
    w2 = np.random.normal(scale=0.5)
    w3 = np.random.normal(scale=0.5)
    w4 = np.random.normal(scale=0.5)
    w5 = np.random.normal(scale=0.5)
    print(w1,w2,w3,w4,w5)
    
    current_bias = 1
    iterations = iterations
    learning_rate = learning_rate
    n = float(len(a))
     
    costs = []
    weight1 = []
    weight2 = []
    weight3 = []
    weight4 = []
    weight5 = []
    previous_cost = None
     
    for i in range(iterations):
         
        y_predicted = (w1 * a) + (w2 * b) + (w3 * c) + (w4 * d) + (w5 * e) + current_bias
        current_cost = mean_squared_error(y, y_predicted)
 
        if previous_cost and abs(previous_cost-current_cost)<=stopping_threshold:
            break
         
        previous_cost = current_cost
 
        costs.append(current_cost)
    
        weight1.append(w1)
        weight2.append(w2)
        weight3.append(w3)
        weight4.append(w4)
        weight5.append(w5)
         
        w1derivative = -(2/n) * sum(a * (y-y_predicted))
        w2derivative = -(2/n) * sum(b * (y-y_predicted))
        w3derivative = -(2/n) * sum(c * (y-y_predicted))
        w4derivative = -(2/n) * sum(d * (y-y_predicted))
        w5derivative = -(2/n) * sum(e * (y-y_predicted))
        
        biasderivative = -(2/n) * sum(y-y_predicted)
         
        w1 = w1 - (learning_rate * w1derivative)
        w2 = w2 - (learning_rate * w2derivative)
        w3 = w3 - (learning_rate * w3derivative)
        w4 = w4 - (learning_rate * w4derivative)
        w5 = w5 - (learning_rate * w5derivative)
        
        current_bias = current_bias - (learning_rate * biasderivative)
                      
     
    plt.figure(figsize = (8,6))
    plt.plot(weight1, costs)
    plt.plot(weight2, costs)
    plt.plot(weight3, costs)
    plt.plot(weight4, costs)
    plt.plot(weight5, costs)
    plt.scatter(weight1, costs, marker='o', color='blue')
    plt.scatter(weight2, costs, marker='o', color='orange')
    plt.scatter(weight3, costs, marker='o', color='green')
    plt.scatter(weight4, costs, marker='o', color='red')
    plt.scatter(weight5, costs, marker='o', color='purple')
    plt.title("Cost vs Weights")
    plt.ylabel("Cost")
    plt.xlabel("Weight")
    plt.show()
     
    return w1,w2,w3,w4,w5, current_bias

def normalize(value):
    norm = np.linalg.norm(value)
    value = value/norm  
    return value
```

## Data Acquisition

```python
a= np.random.normal(scale=10, size=(20,))

b= np.random.normal(scale=10, size=(20,))

c= np.random.normal(scale=10, size=(20,))

d= np.random.normal(scale=10, size=(20,))

e= np.random.normal(scale=10, size=(20,))

Y = np.random.normal(scale=10, size=(20,))
```

```python
print(a)
print()
print(b)
print()
print(c)
print()
print(d)
print()
print(e)
print()
print(Y)
```

## Calling the function and checking the difference in weights

```python
w1,w2,w3,w4,w5, eatimated_bias = gradient_descent(a,b,c,d,e, Y, iterations=10)
print(f"Estimated Weight 1: {w1}\nEstimated Weight 2: {w2}\nEstimated Weight 3: {w3}\nEstimated Weight 4: {w4}\nEstimated Weight 5: {w5}\nEstimated Bias: {eatimated_bias}")
 
Y_pred = w1*a + eatimated_bias
 
plt.figure(figsize = (8,6))
plt.scatter(a, Y, marker='o', color='red')
plt.scatter(b, Y, marker='o', color='blue')
plt.scatter(c, Y, marker='o', color='green')
plt.scatter(d, Y, marker='o', color='yellow')
plt.scatter(e, Y, marker='o', color='black')

plt.plot([min(a), max(a)], [min(Y_pred), max(Y_pred)], color='red',markerfacecolor='red',markersize=10,linestyle='dashed')
plt.plot([min(b), max(b)], [min(Y_pred), max(Y_pred)], color='blue',markerfacecolor='blue',markersize=10,linestyle='dashed')
plt.plot([min(c), max(c)], [min(Y_pred), max(Y_pred)], color='green',markerfacecolor='green',markersize=10,linestyle='dashed')
plt.plot([min(d), max(d)], [min(Y_pred), max(Y_pred)], color='yellow',markerfacecolor='yellow',markersize=10,linestyle='dashed')
plt.plot([min(e), max(e)], [min(Y_pred), max(Y_pred)], color='black',markerfacecolor='black',markersize=10,linestyle='dashed')

plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## Normalizing the data

```python
a = normalize(a)
b = normalize(b)
c = normalize(c)
d = normalize(d)
e = normalize(e)
Y = normalize(Y)
```
