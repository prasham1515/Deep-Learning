```python
from happytransformer import HappyGeneration, GENSettings
happy_gen = HappyGeneration("GPT-NEO", "EleutherAI/gpt-neo-125M")
```

```python
import pandas as pd

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv("final.csv")

# Convert the DataFrame to the desired string format
output_string = ""
for index, row in df.iterrows():
    output_string += f"Keywords: {row['keywords']}\nOutput: {row['text']} \n###\n"
```

```python
len(output_string)
```

```python
keywords=['water bottle, strong, durable, cheap']
```

```python
def create_prompt(training_cases, keywords):
  keywords_string = ", ".join(keywords)
  prompt = training_cases + "\nKeywords: "+ keywords_string + "\nOutput:"
  return prompt
```

```python
prompt = create_prompt(output_string[:-2], keywords)
```

```python
len(prompt)
```

```python
args_beam = GENSettings(num_beams=5, no_repeat_ngram_size=2, early_stopping=True, min_length=1, max_length=50)
result = happy_gen.generate_text(prompt, args=args_beam)
print()
print(result)
```

```python
spl_word = '\n'
res = result.text.partition(spl_word)[0]
# print result
print(res)
```
